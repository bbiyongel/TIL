# Kmean clustering 

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
X = iris.data[:, :2]
y = iris.target
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('length', fontsize=10)
plt.ylabel('width', fontsize=10)
```

```python
km = KMeans(n_clusters=3, n_jobs=4, random_state=21)
km.fit(X)
```

```python
centers = km.cluster_centers_
print(centers)
```

```python
new_labels = km.labels_
#
fig, axes = plt.subplots(1,2, figsize=(16,8))
axes[0].scatter(X[:,0], X[:, 1], c=y, cmap='gist_rainbow', edgecolor='k', s=150)
axes[1].scatter(X[:,0], X[:, 1], c=new_labels, cmap='jet', edgecolor='k', s=150)
axes[0].set_xlabel('length', fontsize=18)
axes[0].set_ylabel('width', fontsize=18)
axes[1].set_xlabel('length', fontsize=18)
axes[1].set_ylabel('width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, color='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, color='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('predict', fontsize=18)
```
